2. What is an interpreter in Python?

Answer- A Python interpreter is a program that runs Python code directly, line by line, without needing to compile it first.

3. What are pre-defined keywords in Python?

Answer - Keywords are special reserved words in Python that have a specific meaning. You cannot use them as variable names, function names, or identifiers.

4. Can keywords be used as variable names?

Answer- No, keywords can't be used as  variable names.

5. What is mutability in Python?

Answer- Mutability means whether an object can be changed after creation. Lists, dictionaries, and sets are mutable; where strings, tuples, and numbers are immutable.

6. Why are lists mutable, but tuples are immutable?

Answer- Lists are mutable so anyone can modify them easily; tuples are immutable to act as fixed, safe, and faster data that can even be used as dictionary keys.

7. What is the difference between “==” and “is” operators in Python?

Answer- "==" checks if values are equal, is checks if two variables point to the exact same object in memory. Use "==" for values, use is only for None, True, False.

8. What are logical operators in Python?

Answer- Python has three logical operators: and (both true), or (at least one true), and not (opposite). They are used to combine or invert conditions.

9. What is type casting in Python?

Answer- Type casting is converting a value from one data type to another using functions like int(), float(), str(), etc. Python can also do it automatically (implicit casting).

10. What is the difference between implicit and explicit type casting?

Answer- Implicit type casting is done automatically by Python (safe), while explicit type casting is done manually by the programmer using functions like int(), float(), str(), etc.

11. What is the purpose of conditional statements in Python?

Answer- The purpose of conditional statements (if, elif, else) in Python is to allow the program to make decisions and execute different blocks of code depending on whether a condition is True or False.

12. How does the elif statement work?

Answer- "elif" lets you check multiple conditions in order — only the first true block runs, and the rest are skipped.

13.  What is the difference between for and while loops?

Answer- "for" loop is used when you know the number of iterations; "while" loop is used when you want to repeat until a condition becomes False.

14. Describe a scenario where a while loop is more suitable than a for loop?

Answer- A "while" loop is more suitable when you don’t know how many times the loop will run and want to continue until a specific condition becomes True or False — for example, asking for a correct password.

```python
6. # Write a function to demonstrate the behavior of mutable and immutable arguments.

def modify_values(x, my_list):
    x = x + 100
    print("Inside function - x (int):", x)

    my_list.append(999)
    my_list[0] = 888
    print("Inside function - my_list:", my_list)


num = 50
numbers = [1, 2, 3, 4]

print("Before function call:")
print("num =", num)
print("numbers =", numbers)

print("\nCalling function...")
modify_values(num, numbers)

print("\nAfter function call:")
print("num =", num)
print("numbers =", numbers)
```

```python
7. # Write a program that performs basic arithmetic operations on two user-input numbers.

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2

if num2 != 0:
    division = num1 / num2
    floor_division = num1 // num2
    modulus = num1 % num2
else:
    division = "Cannot divide by zero"
    floor_division = "Cannot divide by zero"
    modulus = "Cannot find remainder with zero"

power = num1 ** num2

print("\n--- Results ---")
print(f"Addition: {num1} + {num2} = {addition}")
print(f"Subtraction: {num1} - {num2} = {subtraction}")
print(f"Multiplication: {num1} × {num2} = {multiplication}")
print(f"Division: {num1} ÷ {num2} = {division}")
print(f"Floor Division: {num1} // {num2} = {floor_division}")
print(f"Modulus (Remainder): {num1} % {num2} = {modulus}")
print(f"Power: {num1}^{num2} = {power}")
```

```python
9. # Write a Python program to convert user input from string to integer, float, and boolean types.
print("=== Type Conversion Demo ===\n")

str_num = input("Enter a number (as text): ")
str_float = input("Enter a decimal number (as text): ")
str_bool = input("Enter True or False (as text): ")

print("\n--- Converting to Different Types ---")

try:
    num_int = int(str_num)
    print(f"Converted to Integer  : {str_num} → {num_int} (type: {type(num_int)})")
except ValueError:
    print(f"Cannot convert '{str_num}' to integer!")

try:
    num_float = float(str_float)
    print(f"Converted to Float     : {str_float} → {num_float} (type: {type(num_float)})")
except ValueError:
    print(f"Cannot convert '{str_float}' to float!")

if str_bool.lower() == "true":
    converted_bool = True
elif str_bool.lower() == "false":
    converted_bool = False
else:
    converted_bool = bool(str_bool)

print(f"Converted to Boolean   : '{str_bool}' → {converted_bool} (type: {type(converted_bool)})")

print(f"Using bool() directly  : bool('{str_bool}') → {bool(str_bool)}")
```

```python
10. # Write code to demonstrate type casting with list elements.
numbers = ["5", "10", "15", "20"]

print("Original list (strings):")
print(numbers)
print()

int_numbers = [int(x) for x in numbers]

print("After converting to integers:")
print(int_numbers)
print()

float_numbers = [float(x) for x in numbers]

print("After converting to float:")
print(float_numbers)
print()

string_numbers = [str(x) for x in int_numbers]

print("After converting back to strings:")
print(string_numbers)
```

```python
14. # Write a program to reverse a string using a while loop.
text = input("Enter a word: ")
reversed_str = ""
i = len(text) - 1

while i >= 0:
    reversed_str += text[i]
    i -= 1

print("Reversed:", reversed_str)
```

```python
15. # Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number: "))

if num < 0:
    print("No factorial for negative numbers!")
else:
    fact = 1
    i = 1
    while i <= num:
        fact *= i
        i += 1
    print(f"Factorial of {num} = {fact}")
```
